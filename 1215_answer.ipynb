```python
# 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Google Colab
from google.colab import files
uploaded = files.upload()

# 
import io
data = pd.read_csv(io.BytesIO(uploaded['japan_20220319_query.csv']))

# datetime
data['time'] = pd.to_datetime(data['time'])
data['date'] = data['time'].dt.date

# 
plt.figure(figsize=(10, 6))
sns.histplot(data['mag'], bins=20, kde=False)
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()

# 
plt.figure(figsize=(10, 6))
sns.histplot(data['depth'], bins=20, kde=False)
plt.title('Distribution of Earthquake Depths')
plt.xlabel('Depth (km)')
plt.ylabel('Frequency')
plt.show()

# 
earthquake_count_by_date = data.groupby('date').size()
plt.figure(figsize=(15, 6))
earthquake_count_by_date.plot(kind='bar')
plt.title('Daily Earthquake Frequency')
plt.xlabel('Date')
plt.ylabel('Number of Earthquakes')
plt.xticks(rotation=90)
plt.show()
```
